```python
# debemos agarrar cada serie (columna del dataframe) y pasarla a array, y esos son los arrays que voy a graficar buscando alguna
# relación entre ambas variables
# una de las series que sí o sí deberá aparecer será la de si llegó en horario o no (cosa que aparece sólo en el dataset de train)
# o sea el y de nuestro modelo será ese, mientras que el x serán las demás columnas, buscando justamente una relación entre 
# una y otra

# se realizará una estudio de correlación lineal con el índice de Pearson con el objetivo de hallar correlación para 
# posteriormente crear un modelo de regresión lineal
# una vez que encontremos una x que se correlacione correcta y altamente con y, vamos a entrenar el modelo de predicción
```

**y = 'Reached.on.Time_Y.N'**

```python
# pasamos la columna de si llegó a tiempo o no a un array de numpy con la intención de que pueda usarse dentro del modelo de 
# regresión lineal
y = data_train['Reached.on.Time_Y.N'].to_numpy()
```

```python
y
```

```python
type(y)
```

**x = 'Mode_of_shipment'**

```python
# para nuestra primer x, vamos a probar la correlación con 'Mode_of_shipment'
x_mode_of_shipment = data_train['Mode_of_Shipment'].to_numpy()
```

```python
x_mode_of_shipment
```

```python
# definidos x e y, vamos a estudiar la covarianza que existe entre ellas
np.cov(x_mode_of_shipment,y)
# para medir esta covarianza hay que darle un número a cada palabra porque no soporta el formato str. La covarianza debe realizarse 
# entre dos int
```

```python
x_mode_of_shipment_list = x_mode_of_shipment.tolist()
```

```python
x_mode_of_shipment_list
```

```python
for index, value in enumerate(x_mode_of_shipment_list):
    if value == 'Flight':
      x_mode_of_shipment_list[index] = 0
    elif value == 'Ship':
      x_mode_of_shipment_list[index] = 1
    else:
      x_mode_of_shipment_list[index] = 2
```

```python
x_mode_of_shipment_list
```

```python
np.cov(x_mode_of_shipment_list,y)
```

```python
np.corrcoef(x_mode_of_shipment_list,y)
coefs.append(np.corrcoef(x_mode_of_shipment_list,y)[0][1])
```
